```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv("data/training_data_fall2024.csv")
train_data.head()
```

```python
# drop holiday and snow columns based on data analysis
train_data = train_data.drop(columns=["holiday", "snow"])
train_data.head()
```

```python
# Add a feature called "day", a binary feature which is 1 if hour_of_day is between 6-22 otherwise 0
train_data["day"] = ((train_data['hour_of_day'] >= 7) & (train_data['hour_of_day'] <= 20)).astype(int)
train_data.head()
```

```python
# Encode snowdepth as binary feature, 1 if there is any snow, 0 otherwise
train_data["snowdepth"] = ((train_data['snowdepth'] > 0)).astype(int)
train_data.head(5)
```

```python
# Make a new feature called "rain", binary feature that is 1 if precip is non-zero, 0 otherwise
train_data["rain"] = ((train_data['precip'] > 0)).astype(int)
train_data = train_data.drop(columns=["precip"])
```

```python
train_data.to_csv("data/preprocessed_dataset.csv", index=False)
```
